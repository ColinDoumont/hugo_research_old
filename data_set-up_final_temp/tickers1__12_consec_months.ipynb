```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv (r'/Users/Colin/Desktop/Research project/Sigma/Data - Final/Final data/Dataset1 (for Tickers1 computation).csv')
df
```

```python
df['datadate'] = pd.to_datetime(df.datadate, format='%d/%m/%Y')

df['day'] = df['datadate'].dt.strftime('%d')
df['month'] = df['datadate'].dt.strftime('%m')
df['year'] = df['datadate'].dt.strftime('%Y')

df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)
```

```python
df = pd.DataFrame(df, columns= ['datadate','tic','prccm','month','year'])
df
```

```python
prevTIC = 'CAD'
prevMonth = 0
prevYear = 0
count = 0
max = 0
consecutive = False
keep = False
list = []

for index, row in df.iterrows():
    
    thisTIC = row[1]
    thisMonth = row[3]
    thisYear = row[4]
    
    if thisMonth == 1:
        if prevMonth == 12 and thisYear == prevYear+1:
            consecutive = True
        else: consecutive = False
    else:
        if prevMonth == thisMonth - 1 and thisYear == prevYear:
            consecutive = True
        else: consecutive = False
    
    if thisTIC == prevTIC and consecutive == True:
        count = count + 1
    else:
        if thisTIC != prevTIC and keep == True:
            list.append(prevTIC)
            keep = False
        count = 1
    
    if (count >= 12):
        keep = True
        
    prevTIC = thisTIC
    prevMonth = thisMonth
    prevYear = thisYear
    
if keep == True:
    list.append(thisTIC)
```
