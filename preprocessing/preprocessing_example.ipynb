```python
# Importing necessary packages.
import pandas as pd
import numpy as np
from IPython.display import display
from preprocessing import *
```

```python
# Creating simple dataset to perform quality check of preprocessing functions.
data = {'datadate':['1970-03-31', '1970-06-30', '1970-03-31', '1970-03-31', '1970-06-30', '1971-03-31', '1971-03-31'],
        'cusip': ['AAPL', 'AAPL', 'TSLA', 'GOOG', 'GOOG', 'MFST', 'AMZN'],
        'stock_price':[1, None, 3, None, 100, None, None]}
df = pd.DataFrame(data)
display(df)
```

```python
# Filling NaN forwardly.
ffill(df, ['stock_price'])
```

```python
# Filling NaN with cross-section median.
mfill(df, ['stock_price'])
```

```python
# Adding lags to dataframe.
df_lagged, X_col_list, y_col_list = lagged(df, ['stock_price'])
display(df_lagged)
print(X_col_list)
print(y_col_list)
```

```python
# Dropping rows with NaN.
df.dropna()
```

```python
# Creating simple dataset to perform quality check of preprocessing functions.
data = {'datadate': ['2015-02-28', '2015-05-31', '2015-8-31', '2015-08-31', '2015-11-30', '2016-02-29', '2016-05-31'],
        'cusip': ['AAPL', 'AAPL', 'AAPL', 'TSLA', 'AAPL', 'AAPL','AAPL'],
        'stock_price':[1, 1, 1, 100, 1, 1, 1],
        'stock_price_10x':[10, 10, 10, 1000, 10, 10, 10]}
df = pd.DataFrame(data)
df['datadate'] = pd.to_datetime(df.datadate, format='%Y-%m-%d')
display(df)
```

```python
# Converting quarterly data to trailing twelve months (TTM) data.
ttm(df, ['stock_price', 'stock_price_10x'])
```
