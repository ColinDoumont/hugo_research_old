```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv (r'/Users/Colin/Desktop/Research project/Sigma/Data - Final/Final data/Dataset2 (for Tickers2 computation).csv')
print(df)
```

```python
# data formatting so that Python can "understand" it

df['datadate'] = pd.to_datetime(df.datadate, format='%d/%m/%Y')

df['month'] = df['datadate'].dt.strftime('%m')
df['year'] = df['datadate'].dt.strftime('%Y')

df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)

df['day'] = 1
#df['month'] += 1
```

```python
df = pd.DataFrame(df, columns= ['datadate','tic','prccq','cshoq','day','month','year'])
df = df.rename(columns = {'datadate': 'Date'})
print(df)
```

```python
# load inflation data
inflation = pd.read_excel (r'/Users/Colin/Desktop/Research project/Sigma/Data - Final/Inflation adjustment/CPIAUCNS.xls')

# create index multiplier
inflation['CPI_Multiplier'] = inflation['CPIAUCNS'].iloc[-1] / inflation['CPIAUCNS']
```

```python
# create new dataframe to store date info
new_df = df[['year', 'month', 'day']]

# create new column 'Date' in original data w/ datetime conversion
df['Date'] = pd.to_datetime(new_df)

# merge dataframe
df = pd.merge(df, inflation, how='left', on='Date')
```

```python
print(df)
```

```python
# compute adjusted market capitalisation by multiplying shares oustanding, closing price and CPI multiplier
df['CPIAdjMarketCap'] = df['prccq'] * df['cshoq'] * df['CPI_Multiplier']
```

```python
df = pd.DataFrame(df, columns= ['tic','CPIAdjMarketCap'])
print(df)
```

```python
# create list of tickers with adjusted market cap above 100 (million)

prevTIC = 'CAD'
keep = False
list = []

for index, row in df.iterrows():
    
    thisTIC = row[0]
    thisMarkCap = row[1]
    
    if thisMarkCap >= 100:
        keep = True
    
    if prevTIC != thisTIC and keep == True:
        list.append(prevTIC)
        keep = False
        
    prevTIC = thisTIC
    
print(len(list))
```

```python
print(*list, sep='\n')
```
